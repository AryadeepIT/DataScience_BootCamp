```python
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')

impute = SimpleImputer(missing_values= np.nan, strategy= 'mean')
impute.fit(df.iloc[ : , 2:3].values)
df.iloc[ : , 2:3] = impute.transform(df.iloc[ : , 2:3].values)

df = df.dropna()
```

```python
df.head()
```

### How many free and paid Apps are there in ART_AND_DESIGN

```python
df_pr = df[df['Category'] == 'ART_AND_DESIGN']
print("There are", len(df_pr[df_pr['Type'] == 'Free']),"free apps in ART_AND_DESIGN category")

print("There are", len(df_pr[df_pr['Type'] == 'Paid']),"paid apps in ART_AND_DESIGN category")
```

### How many apps having rating more than 4.5 in ART_AND_DESIGN

```python
df_pr = df[df['Rating'] > 4.5]
print("There are", len(df_pr[df_pr['Category'] == 'ART_AND_DESIGN']),"apps in ART_AND_DESIGN category having rating > 4.5")
```

### How many apps are there in family category with rating more than 4.5 and free

```python
df_pr = df[df['Rating'] > 4.5]
df_ct = df_pr[df_pr['Category'] == 'FAMILY']
print("There are", len(df_ct[df_ct['Type']== 'Free']),"free apps in FAMILY category having rating > 4.5")
```

### List all the free apps with rating more than 4.5 and category FAMILY

```python
print("-"*100)
print("These are the free apps where category is family with rating more than 4.5")
print("-"*100)

df_pr = df[df['Rating'] > 4.5]
df_ct = df_pr[df_pr['Category'] == 'FAMILY']
result = df_ct[df_ct['Type']== 'Free'].values
for i in result:
    print(i[0])

print("-"*100)
```
