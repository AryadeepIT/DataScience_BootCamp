### Analysing India

```python
df_pr = df[df['Country Name'] == 'India']
df_pr.plot(kind = 'line', 
           x = 'Year', 
           y = 'Value', 
           figsize = (15, 6), 
           grid = True,
           legend = False, 
           ylabel = 'GDP',
           xlabel = 'YEARS')
```
