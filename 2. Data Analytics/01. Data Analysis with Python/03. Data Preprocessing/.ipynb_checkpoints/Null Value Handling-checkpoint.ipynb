### Filling the values  - Numeric   (Using Imputers)

```python
# calculate average numeric value and fill
# age and salary has only numeric values

imputer = SimpleImputer(missing_values = np.nan, strategy = 'mean')
```

```python
df.iloc[:,1:3].values
```

```python
x = imputer.transform(df.iloc[:, 1:3].values)
```

```python
x
```

```python
df.iloc[:, 1:3] = x
```

```python
df
```

```python
# calculate average numeric value and fill
# age and salary has only numeric values

imputer = SimpleImputer(strategy='mean', missing_values=np.nan)


imputer.fit(df.iloc[:, 1:3].values)
df.iloc[:, 1:3] = imputer.transform(df.iloc[:, 1:3].values)
```

```python
# new dataframe

df
```

## Categorical

```python
imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)


imputer.fit(df.iloc[:, :1].values)
df.iloc[:, :1] = imputer.transform(df.iloc[:, :1].values)
```

```python
df
```

```python
imputer = SimpleImputer(strategy='most_frequent', missing_values=np.nan)


imputer.fit(df.iloc[:, 3:].values)
df.iloc[:, 3:] = imputer.transform(df.iloc[:, 3:].values)
```

```python
df
```
