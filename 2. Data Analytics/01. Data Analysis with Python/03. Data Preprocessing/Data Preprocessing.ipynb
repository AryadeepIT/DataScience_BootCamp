### Removing null values

```python
import pandas as pd
```

```python
data = pd.read_csv('googleplaystore.csv')
```

```python
data.head(10)
```

```python
data.isnull()   # to check if there is any null value
```

```python
data.isnull().sum()    # check total null value in respective columns
```

```python
data.dropna()   # to drop null values rows
```

```python
df = data.dropna()
```

```python
df.isnull().sum()  # now there are no null values
```
